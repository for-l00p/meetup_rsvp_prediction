<i>Scraping member profile details using BeautifulSoup.</i>

The meetup API does not have an easy way to get a list of all groups that each individual member belongs to. However, they do provide a member ID which can be used to look up with member profile online. On that page, if shared, we can get a list of all meetup groups that each individual belongs to (see my meetup page below). I will use BeautifulSoup to get the list of groups on each page as well as individual interests if provided.

### Import libraries

```python
import selenium
from bs4 import BeautifulSoup
import sqlite3
import pandas as pd
import requests
from meetup_api_functions import get_member_profile
import time
import pickle
```

### Load in all the data

```python
# get a count of the total number of meetup members in NYC
url="https://api.meetup.com/2/cities"
response = requests.get(url, params = {'query': 10036})
member_count = response.json()['results'][0]['member_count']
print("There are {} meetup members in New York City.".format(member_count))
```

```python
# open up the meetup database to get member_urls
connection = sqlite3.connect('meetup_database.db')
cursor = connection.cursor()
```

```python
member_urls = cursor.execute('SELECT member_url FROM nyc_members;').fetchall()
print(len(member_urls))
unique_member_urls = list(set(member_urls))
print(len(unique_member_urls))
```

```python
# currently the urls are listed in a tuple; let's remove them from tuple set and into a list of strings
unique_member_urls = [m[0] for m in unique_member_urls]
```

```python
# this looks much better
unique_member_urls[0]
```

```python
print(f"We will scrape {len(unique_member_urls)} unique members.")
```

```python
# run a for loop to scrape member info
# can stop the scraping early once enough data has been collected
all_profiles = []
total_urls = len(unique_member_urls)

for i, url in enumerate(unique_member_urls):
    try:
        if i in range(1000, 17000, 1000):
            all_profiles.append(get_member_profile(url))
            with open(f'member_profiles_{i}.pkl', 'wb') as f: 
                    pickle.dump(all_profiles, f)
        else:
            all_profiles.append(get_member_profile(url))
    except Exception as e:
        print(e)
    time.sleep(1)
    print("On {}/{}".format(i+1, total_urls))
        
# saving all member profiles
with open(f'all_member_profiles.pkl', 'wb') as f: 
    pickle.dump(all_profiles, f)
```

```python
# getting back member_profiles_16000:
with open('member_profiles_16000.pkl', 'rb') as f:
    member_profiles_16000 = pickle.load(f)
```

```python
print(f"Scraped {len(member_profiles_16000)} profiles")
```
